```r
# find flights with an arrival delay of 120 or more minutes.

filter(flights, arr_delay >= 120)
```

```r
# Find flights where destination is either “IAH” or “HOU”.

filter(flights, dest == "IAH" | dest == "HOU")
# filter(flights, dest %in% c("IAH", "HOU"))
# 
```

```r
unique(flights$carrier)
```

```r
## flights are those that departed in months 7 (July), 8 (August), and 9 (September).

filter(flights, month >= 7, month <= 9)
# filter(flights, month %in% 7:9)
# filter(flights, month == 7 | month == 8 | month == 9)
```

```r
# Flights that arrived more than two hours late

filter(flights, arr_delay > 120)
```

```r
# Finding flights that departed between midnight and 6 a.m.
# In dep_time, midnight is represented by 2400

filter(flights, dep_time <= 600 | dep_time == 2400)
```

```r
# Flights departed  in summer

filter(flights, between(month, 7, 9))  # Here Month 
```

```r
filter(flights, origin == "EWR" | origin == "JFK" | origin == "LGA")
```

```r
aggregate(flights$dep_delay, by=list(origin=flights$origin), FUN=mean)
```

```r
data = flights %>% group_by(origin) %>% summarise_at(vars(dep_delay),funs(mean(.,na.rm=TRUE)))

ggplot(data=data, aes(x=origin, y=dep_delay,fill=origin))+ geom_bar(stat="identity") + ylab("Avg_Dep_Delay") +  ggtitle("Average Departure Delay By Airport")
```

```r
ggplot(data=data, aes(x=origin, y=dep_delay,fill=origin))+ geom_bar(stat="identity") + ylab("Avg_Dep_Delay") +  ggtitle("Average Departure Delay By Airport")
```

```r
df= select(flights,c('month',arr_delay))

df1 = df %>% group_by(month) %>% summarise_at(vars(arr_delay),funs(max(.,na.rm=TRUE)))

ggplot(data=df1, aes(x=month, y=arr_delay,fill=arr_delay))+ geom_bar(stat="identity") + labs(title="Temperatures\n") + scale_x_continuous(breaks=df1$month,limits=c(0, 12)) + ylab("Max_Arrival_Delay") +   ggtitle("Maximum Time of Arrival Delay by Month")
```
