### Scatter plot

```r
# Syntax : plot(y ~ x)
plot(qsec ~ mpg, data=mtcars,pch=16, cex=2, col="red")
```

```r
plot(x = mtcars$qsec,mtcars$mpg,pch=16, cex=2, col="red")
```

```r
plot(qsec ~ mpg, data=mtcars, pch=16, cex=2, col=c("red","green"))

legend("topleft", pch="*", col=c("blue", "green"), c("A", "B"), cex=0.8,title="Celltype")
```

```r
with(mtcars, plot(mpg, qsec,pch=16, cex=2, col="red"))
```

### boxplot

```r
boxplot(mtcars$mpg, col="green") # One Dimentional
```

```r
# Two-Dimensional Plotting for box plotting
boxplot(mpg~gear, data=mtcars, col = "green")  # boxplot(y~x)
```

### Histograms

```r
hist(mtcars$mpg, col = "green")               ## Plot 1
hist(mtcars$mpg, col = "green", breaks = 25)  ## Plot 2
hist(mtcars$mpg, col = "green", breaks = 50)  ## Plot 3
```

```r
hist(subset(mtcars, cyl == 4)$mpg, col = "yellow")        ## Plot 1
hist(subset(mtcars, cyl == 8)$mpg, col = "yellow")        ## Plot 2"
```
