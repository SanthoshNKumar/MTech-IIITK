#  Slicing with logical vectors

```r
df = read.csv("sampledata//vehicle.csv")

head(df)
```

```r
# Create a new df with only the rows of df
#  where State equals CA

df.CA <- df[df$State == "CA", ]

head(df.CA)
```

```r
# Create a new df with only the rows of df
#  where supp equals OJ and dose < 1

df.CA.lc <- df[df$State == "CA" & df$lc <100, ]

head(df.CA.lc)
```

##  Slicing with subset()

```r
subset(x = df, subset = lc < 100 & State == "CA" & lh >= 1)
```
