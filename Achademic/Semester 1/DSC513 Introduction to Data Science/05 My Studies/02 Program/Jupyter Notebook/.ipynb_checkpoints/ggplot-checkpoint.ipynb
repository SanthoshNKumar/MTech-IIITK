```r
# GG : Grammer Of Graphics  
# ggplot2 is a system for declaratively creating graphics, based on The Grammar of Graphics.

# aes : aesthetics
#    - The data is to map onto the Aesthetics attributes such as x-axis, y-axis, color, fill, size, labels, alpha, shape, 
#      line width, line type
```

```r
library(ggplot2)
library(dplyr)
```

```r
head(mtcars)
```

```r
ggplot(data = mtcars, aes(x = hp, y = mpg, col = disp)) + geom_point()
```

```r
ggplot(data = mtcars,aes(x = hp, y = mpg, size = disp)) + geom_point()
```

```r
ggplot(data = mtcars, aes(x = hp, y = mpg, col = factor(cyl),shape = factor(am))) + geom_point()
```

```r
# Histogram plot
ggplot(data = mtcars, aes(x = hp)) + geom_histogram(binwidth = 5)
```

## Facet Layer

```r
p <- ggplot(data = mtcars, aes(x = hp, y = mpg, shape = factor(cyl))) + geom_point()
```

```r
# Separate rows according to transmission type
p + facet_grid(am ~ .)
```

```r
# Separate columns according to cylinders
p + facet_grid(. ~ cyl)
```

```r

ipeds %>% group_by(Sector) %>% summarise(c=unique(length(Name))) %>%
    ggplot(aes(Sector,c,fill=Sector)) + geom_bar(stat="identity",,alpha=0.5) + theme_minimal() + 
        theme(axis.text.x = element_text(angle=90),legend.position = "bottom") + 
            labs(title="Universities by Sectors")  
```
