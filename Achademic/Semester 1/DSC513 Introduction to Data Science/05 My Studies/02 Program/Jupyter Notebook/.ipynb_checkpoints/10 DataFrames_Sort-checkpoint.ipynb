# order

```r
head(data[order(data$lh, decreasing = TRUE),])
```

```r
head(data[order(data$lh, data$fm, decreasing = TRUE),])
```

```r
head(data[order(data$lh, decreasing = FALSE),])
```

```r
head(data[order(data$State, decreasing = TRUE),])
```

```r
head(data[order(data$State, decreasing = FALSE),])
```

```r
# Sort by fm(descending) and lc(acending)

head(data[order(-data$fm, data$lc),])
```

# arrange() Function from dplyr

```r
library("dplyr")  

head(arrange(data, State))
```

```r
# Ascending sort of variable lh
head(arrange(data,lh))
```

```r
# Ascending sort of variable lh and fm

head(arrange(data,lh,fm))
```

```r
# Descending sort of variable lh and ascending sort of lh

head(arrange(data,desc(lh),fm))
```

# Setorder() from data.table

```r
library("data.table")

setorder(data,State)
```
