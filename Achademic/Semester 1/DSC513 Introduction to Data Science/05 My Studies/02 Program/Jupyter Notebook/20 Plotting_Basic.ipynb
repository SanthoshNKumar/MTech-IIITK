# Color 

```r
x <- rnorm(500)
y <- x + rnorm(100)

par(mfrow=c(2,2))

plot(x, cos(x),col= 'red')
plot(x, sin(x),col = 'blue')
plot(x, cos(x),col = 'yellow')
plot(x, sin(x),col = 'pink')
```

```r
plot(x, y, main = "Title", sub = "Subtitle",
    pch  = 16,
    col = "red",            # Symbol color
    col.main = "green",     # Title color
    col.sub = "blue",       # Subtitle color
    col.lab = "sienna2",    # X and Y-axis labels color
    col.axis = "maroon4",   # Tick labels color
    fg = "orange")          # Box color
```

```r
group <- ifelse(x < 0 , "car", ifelse(x > 1, "plane", "boat"))

df <- data.frame(x = x, y = y, group = factor(group))

# Color by group
plot(df$x, df$y, col = df$group, pch = 16)

# Change group colors
colors <- c("red", "green", "blue")

#plot(df$x, df$y, col = colors[df$group], pch = 16)

# Change color order, changing levels order
#plot(df$x, df$y, col = colors[factor(group, levels = c("car", "boat", "plane"))],pch = 16)
```

# Plot Type

```r
# Changing Color and Plot Type

"p" - points
"l" - lines
"b" - both points and lines
"c" - empty points joined by lines
"o" - overplotted points and lines
"s" and "S" - stair steps
"h" - histogram-like vertical lines
"n" - does not produce any points or lines
```

```r
par(mfrow=c(2,3))

plot(x, cos(x),type = 'p')
plot(x, sin(x),type = 'l')
plot(x, cos(x),type = 'b')
plot(x, sin(x),type = 'c')
plot(x, cos(x),type = 'o')
plot(x, sin(x),type = 's')
```

## Change axis scale in R

```r
# New data to avoid negative numbers
s <- 1:25
u <- 1:25

par(mfrow = c(2, 2))

# Default
plot(s, u, pch = 19,main = "Untransformed")

# Log scale. X-axis
plot(s, u, pch = 19, log = "x", main = "X-axis transformed")

# Log scale. Y-axis
plot(s, u, pch = 19, log = "y",    main = "Y-axis transformed")

# Log scale. X and Y axis
plot(s, u, pch = 19, log = "xy", main = "Both transformed")
```

```r
par(mfrow = c(1, 3))

# Log-log
plot(log(s), log(u), pch = 19, main = "log-log")

# log(x)
plot(log(s), u, pch = 19, main = "log(x)")

# log(y)
plot(s, log(u), pch = 19,main = "log(y)")

par(mfrow = c(1, 1))
```

## Multi plotting

```r
#Set space for 2 rows and 3 columns.
par(mfrow=c(2,3))

#Plot out the graphs using various options.
plot(x,cos(x),col=c('blue','orange'),type='o',pch=19,lwd=2,cex=1.5)

plot(x,x*2,col='red',type='l')

plot(x,x^2/3+4.2, col='violet',type='o',lwd=2,lty=1)

plot(c(1,3,5,7,9,11),c(2,7,5,10,8,10),type='o',lty=3,col='pink',lwd=4)

plot(x<-seq(1,10,0.5),50*x/(x+2),col=c('green','dark green'),type='h')

plot(x,log(x),col='orange',type='s')
```

```r
# main : title 
# lty : line type
# lwd : line width
# col : color
# col.main : Color of Main title 
# type : 
# xlab : x axis label
# ylab : y axis label
```

```r
#Displaying the title with color

x1<- c(1,3,5,7,9,11)
y1 <- c(2,7,5,10,8,10)

labelset <-c('one','three','five','seven','nine','eleven')

plot(x1,y1,type='o',lty=3, col='pink',lwd=4,main="This is a graph",col.main='blue',
     xlab="Time",ylab="Performance")

text(x1+0.5,y1,labelset,col='red')

legend('topleft',inset=0.05,"Performace",lty=3,col='pink',lwd=4)
```

# Overlaying Graphs

```r
x=seq(2,10,0.1)

y1=x^2
y2=x^3

plot(x,y1,type='l',col='red')

lines(x,y2,col='green')

abline(v=c(4,6,8),col="dark green",lty=2) # vertical line
abline(h=c(20,40,60),col="dark green",lty=2) # horizontal line

legend('bottomright',inset=0.05,c("Squares","Cubes"),lty=1,col=c("red","green"),title="Graph type")

```
